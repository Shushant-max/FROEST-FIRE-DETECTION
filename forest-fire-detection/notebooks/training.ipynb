## Data Preparation

Make sure your training data is organized as follows:
- `data/raw/fire/` - Images containing fire
- `data/raw/no_fire/` - Images without fire

Each folder should contain JPG/PNG images.

````python
# Check data availability
fire_dir = '../data/raw/fire'
no_fire_dir = '../data/raw/no_fire'

if os.path.exists(fire_dir):
    fire_count = len([f for f in os.listdir(fire_dir) if f.endswith(('.jpg', '.jpeg', '.png'))])
    print(f"Fire images found: {fire_count}")
else:
    print("Fire images directory not found. Please add training data.")

if os.path.exists(no_fire_dir):
    no_fire_count = len([f for f in os.listdir(no_fire_dir) if f.endswith(('.jpg', '.jpeg', '.png'))])
    print(f"No-fire images found: {no_fire_count}")
else:
    print("No-fire images directory not found. Please add training data.")
````
